```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,5,11)
y = x**2
```

Graficamos x & y

```python
plt.plot(x,y)
plt.show()
```

Modificacion del color de la grafica  


```python
plt.plot(x,y, 'g')
plt.show()
```

### Cambiar más parámetros
Como se muestran la linea  
  


```python
plt.plot(x,y, 'yD:') #grafica de color amarillo, con diamantes y puntos consecutivos
plt.show()
```

#### Crear un histograma

```python
plt.hist(y)  #histograma de los valores de y
plt.show()
```

## Grafico de Pie 

```python
plt.pie(y)
plt.show()
```

### Conocer correlación entre variables

```python
plt.scatter(x,y)  #correlacion entre la variable x e y
plt.show()
```

### Distribución de los datos  
* Distribución de los datos: Gráfico de caja

```python
plt.boxplot(x)
plt.show()
```
